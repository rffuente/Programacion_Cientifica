### Pregunta 1

El formato que utiliz√≥ para crear el archivo HDF5 fue sin _chuncks_, debido a que utilizandolos la creacion del dataset se demoraba mucho mas que hacerlo sin _chunks_. Adem√°s, personalmente no manejo tan bien el uso de esta herramienta, por lo que se me hace mas facil ocupar lo visto en clases.  

````python
#Creamos un archivo HDF5, donde como parametros le pasamos el tama√±o total de la imagen = (8660, 10260, 3)
f = h5py.File("actividad_3.h", "w")
dset = f.create_dataset("dataset", (img.shape[0],img.shape[1],img.shape[2]), dtype='uint8')
````

````python
#Llenamos nuestro archivo con los datos de img, separandolo por canal
for i in range(img.shape[0]):
    tmp = img[i,:,0]
    dset[i,:,0] = tmp
    del tmp
    
for i in range(img.shape[0]):
    tmp = img[i,:,1]
    dset[i,:,1] = tmp
    del tmp
    
for i in range(img.shape[0]):
    tmp = img[i,:,2]
    dset[i,:,2] = tmp
    del tmp
    
#Borramos img de la memoria principal.   
del img
````

### Pregunta 2

Lo que haremos sera crear una lista para guardar los valores de las operaciones por fila. Luego por fila, crearemos submatrices de 5x5, las que multiplicaremos por el kernel y obtendremos un valor, guardando este en una lista. Haremos esto por canal, por lo que obtendremos una lista de $\displaystyle \left(\frac{8660}{5} \cdot \frac{10260}{5}\right)$ elementos.

````python
#Creamos una funcion para aplicar el filtro
def apply_filter(dset,K,canal):
    #Lo que haremos sera crear una lista para guardar los valores de las operaciones por fila.
    #Luego por fila, crearemos submatrices de 5x5, las que multiplicaremos por el kernel y obtendremos un valor,
    #guardando este en una lista. Haremos esto por canal, por lo que obtendremos una lista de (8660/5)*(10260/5) elementos.
    list_values_filter = []
    #haremos que el for haga saltos de 5x5.
    for i in range(0,dset.shape[0],5):
        
        sub_matrix = np.zeros((5,5))
        sub_row = dset[i:i+5,:,canal]
        
        for j in range(3,sub_row.shape[1]-1,5):
            sub_matrix = sub_row[:,j-3:j+2]
            list_values_filter.append((K.ravel() * sub_matrix.ravel()).sum())
            #borramos estas variables de la memoria
            del sub_matrix
            
        del sub_row
    return np.array(list_values_filter)
````

Ejecutamos el _timeit_ y _memmit_ de cada funcion y por cada canal: 

* ## timeit:

````python
#Kernel de filtro
K = 1/25. * np.ones((5,5))
````

````python
%%timeit
r = apply_filter(dset,K,0)
````

````python
%%timeit
g = apply_filter(dset,K,1)
````

````python
%%timeit
b = apply_filter(dset,K,2)
````

* ## memmit:

````python
%%memit
r = apply_filter(dset,K,0)
````

````python
%%memit
g = apply_filter(dset,K,1)
````

````python
%%memit
b = apply_filter(dset,K,2)
````

Creamos una funci√≥n para reconstruir los 3 canales, llamando a canda funci√≥n y haciendole un _reshape_ para convertirla a matriz. Luego, creamos una matriz que almacenar√° estas matrices y guardamos en cada canal la matriz correspondiente a su color (R-G-B), devolviendo la matriz reconstruida con sus 3 canales:

````python
def Rebuild_image(dset,K):
    r = apply_filter(dset,K,0).reshape((1732, 2052))
    g = apply_filter(dset,K,1).reshape((1732, 2052))
    b = apply_filter(dset,K,2).reshape((1732, 2052))

    image_reconstructed = np.zeros((r.shape[0],r.shape[1],3),'uint8')

    image_reconstructed[:,:,0] = r
    image_reconstructed[:,:,1] = g
    image_reconstructed[:,:,2] = b
    
    return image_reconstructed
````

Basicamente el filtro aplica un promedio a cada submatriz de 5x5 de la matriz original, sin hacer _overlaping_ entre submatrices, realizando el producto punto el kernel del filtro por cada submatriz. 

### Pregunta 3

Finalmente mostramos la imagen reconstruida:

````python
image = Rebuild_image(dset,K)
image_plot(image)
````

Nos fijamos que la imagen pierde algo de calidad, debido a que su tama√±o se redujo bastante en comparaci√≥n a la imagen original.
